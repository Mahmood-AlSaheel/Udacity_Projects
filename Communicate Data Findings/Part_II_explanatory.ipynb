## Dataset Overview and Executive Summary

>The dataset contains information about rides made in a bike sharing system covering the greater San Francisco Bay area.



```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('201902-fordgobike-tripdata.csv')
df.head()
```

## (Users Types)

> What are the differences in bike usage frequency and patterns among male, female, and other gender categories within the bike-sharing program?


```python

plt.figure(figsize=(4, 6))
sns.countplot(data=df, x='member_gender')

def plot_la(title, x, y):
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)

title='Users Types'
x='Users'
y='Counts'
plot_la(title,x,y)
```


It is obvious from the data shown in the chart that the majority of bike users are male more than 120000 users.

## (Member Birth by bike share for all trips)

What is the extent and nature of outliers in the relationship between participation in the 'Bike Share for All Trip' program and by birth year of members?

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='bike_share_for_all_trip', y='member_birth_year', data=df,palette='Set2')
plt.title('Member Birth by bike share for all trips')
plt.xlabel('Bike share_for all trip?')
plt.ylabel('Birth year');
```

The chart shows how the birth years of users differ between those who participate in the bike share System and those who don't,also we can notice that most min outliers are between 1940 birth years and 1980, and there's no max outliers.

## (Distribution of Bike Duration)

What is the typical duration of bike usage in seconds among participants of the bike-sharing system?

```python
fig=plt.figure(figsize=[8,6])
bins= 7 ** np.arange(0.5+0,3, 0.1)
plt.hist(data = df, x = 'duration_sec', bins=bins)
plt.xlabel('Duration in Seconds')
plt.ylabel('Counts')

plt.xscale('log')
plt.title('Bike Duration');
```

From the  chart above we can notice most users spends duration of bike trips are more than 10^2 after scale it to logarithmic scale by using plt.xscale('log'),the distribution of histogram goes skewed left.
